```python
import pandas as pd
```

### 1.딕셔너리를 이용하여 data frame 생성

```python
score={'학번':[201524213,201621223,201621233, 201611111, 201722222],
       '이름':['오지환','박용택','정성훈', '다니엘', '지민'],
       '학과':['경영정보','경영정보','경영정보','컴퓨터공학','컴퓨터공학'],
       'Quiz':[15,20,10,17,18],
       '중간고사':[20,20,15,18,16]}   # key와 value를 입력해주어 딕셔너리 생성

df= pd.DataFrame(score)  # 데이트프레임으로 변경
df
```

```python
df['학번']    # 학번 column 추출 
```

### 2. 위에서 생성한 데이터프레임으로부터 인덱싱을 이용하여 이름과 중간고사를 추출하시오.

```python
df[['이름', '중간고사']]    # 이름과 중간고사 coulmn 추출
```

### 3. reindex를 이용하여 2, 0, 4, 1, 3 순으로 순서를 변경하시오.

```python
df = df.reindex([2, 0, 4, 1, 3])   # reindex -> 인덱스 변경
df
```

### 4. iloc과 loc을 각각 이용하여 박용택 학생의 record를 추출하시오. (iloc으로 한번, loc으로 한번 수행)

```python
print(df.iloc[3])   # iloc를 통해 위치상 4번째 행 추출(박용택)
print(df.loc[1])    # loc를 통해 인덱스가 1인 행 추출
```

### 5. iloc과 loc을 각각 이용하여 모든 학생의 이름과 중간고사 점수를 추출하시오. (iloc으로 한번, loc으로 한번 수행)

```python
print(df.iloc[:, [1, 4]])             # 행 = 모든 행, 열을 필터할 때 숫자로 
print(df.loc[:, ['이름', '중간고사']])   # 열을 필터할 때 컬럼명으로 
```

### 6. 중간고사 점수의 평균을 구하시오.

```python
df['중간고사'].mean()  # mean() -> 평균
```

### 7. 중간고사 점수가 가장 낮은 학생의 index를 구하시오.

```python
df['중간고사'].idxmin()  # 중간고사에서 min인 index 
```

### 8. 퀴즈 점수가 가장 높은 학생의 학번을 구하시오.


```python
df.loc[df['Quiz'].idxmax(), '학번']
# 퀴즈 점수가 max인 index인 행의 학번 추출
```

### 9. 위 데이터프레임의 각 열에 대해서 개수, 평균, 표준편차, 최소최대값 등의 다양한 통계량을 출력하시오.

```python
df.describe()
```

### 10. 위 데이터프레임을 중간고사 점수에 따라 내림차순으로 정렬하시오.

```python
df.sort_values('중간고사', ascending=False) # False = 내림차순
```

### 11. 학생 중에서 퀴즈 점수가 평균 이상인 학생들의 이름을 출력하시오.

```python
quiz_mean = df['Quiz'].mean()
df[df['Quiz'] > quiz_mean]['이름'] # 평균 이상인 학생의 이름 추출
```

### 12. 학생 중에서 퀴즈와 중간고사 점수 모두 평균 이상인 학생들의 record를 출력하시오.

```python
quiz_mean = df['Quiz'].mean()
mid_mean = df['중간고사'].mean()
df[(df['Quiz'] > quiz_mean) & (df['중간고사'] > mid_mean)] # & = 두조건 모두 만족해야 추출
```

### 13. value_counts를 이용하여 학과 별로 학생의 수를 구하시오.

```python
df['학과'].value_counts()    # 학과 value count
```

### 14. groupby를 이용하여 학과 별로 중간고사의 평균을 구하시오.

```python
df.groupby('학과')['중간고사'].mean() # groupby = 학과별로 묶음 
```
